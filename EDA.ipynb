```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country
```

```python
#combining both the dataframes
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
#checking just Dtypes
final_df.dtypes
```

```python
country_names = final_df['Country'].value_counts() #it shows the max number of transactions happenining are in india
country_names
```

```python
#plotting the pie chart
#through pie chart, we want to see which country has the max transactions
plt.pie(country_names.values, labels=country_names.index);
```

```python
#selecting only top 3 countries that uses zomato
plt.pie(country_names.values[:3], labels=country_names.index[:3], autopct = '%1.3f%%');
```

<b> Obervations: </b> We can see max number of transactions are from India. And then United States Precedes then the United Kingdom

## Working on Numerical Variables

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

## Observations

1- From the above output we can see that, when the rating is in between 4.5 to 4.9 it will be <b> excellent </b> <br>
2- When the rating is in between 4.0 to 4.4, it is considered as <b> Very Good </b><br>
2- When the rating is in between 3.5 to 4.0, it is considered as <b> Good </b><br>
2- When the rating is in between 2.5 to 3.4, it is considered as <b> Average </b><br>

```python
import matplotlib 
matplotlib.rcParams['figure.figsize'] = (10,6)
sns.barplot(x = ratings['Aggregate rating'], y = ratings['Rating Count'])
```

```python
#coloring the ratings wrt to the colors provided
sns.barplot(x = ratings['Aggregate rating'], y = ratings['Rating Count'],hue = ratings['Rating color'],
            palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Observations

1. Not Rating Count is very High
2. Max number of Ratings are in between 2.5 and 3.4, wrt to color spread

```python
sns.countplot(x=ratings['Rating color'],palette = ['purple', 'red', 'orange', 'yellow', 'green', 'green'])
```

In the above chart, we can now see that the Orange has the highest rating. So this is wrt to frequencies in the DF, not the rating count
